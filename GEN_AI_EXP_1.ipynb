```python
from transformers import pipeline
```

```python
generator = pipeline(
    "text-generation",
    model="gpt2"
)

print("Model Loaded Successfully!")
```

```python
headline = "India launches new AI policy to boost innovation"

result = generator(
    headline,
    max_length=120,
    num_return_sequences=1,
    temperature=0.8
)

print(result[0]["generated_text"])
```
